**(a)** Use the numerical solution to see what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs. t for 5 different initial angles, $θ_0$ , spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4? What happens to the trajectory when the initial angle is 0 or π ?

```python
def deriv(theta,t, g, ell):
    return np.array([ theta[1], \
                     -1.0*(g/ell)*np.sin(theta[0]) ])
# (omega = 0 = dtheta, theta = domega) 
```

```python
g = 9.81
ell = 0.5
omega0 = 0.0
th0 = np.array([0.0, 45.0*np.pi/180., 90.0*np.pi/180., 135.0*np.pi/180., 180.0*np.pi/180.])
t = np.arange(0.,10.+0.01,0.01)

theta_approx0 = lambda t: th0[0]*np.cos(np.sqrt(g/ell)*t)
theta_approx1 = lambda t: th0[1]*np.cos(np.sqrt(g/ell)*t)
theta_approx2 = lambda t: th0[2]*np.cos(np.sqrt(g/ell)*t)
theta_approx3 = lambda t: th0[3]*np.cos(np.sqrt(g/ell)*t)
theta_approx4 = lambda t: th0[4]*np.cos(np.sqrt(g/ell)*t)

theta_init0 = np.array([th0[0],omega0])
theta_init1 = np.array([th0[1],omega0])
theta_init2 = np.array([th0[2],omega0])
theta_init3 = np.array([th0[3],omega0])
theta_init4 = np.array([th0[4],omega0])

theta0 = odeint(deriv, theta_init0, t, args=(g, ell))
theta1 = odeint(deriv, theta_init1, t, args=(g, ell))
theta2 = odeint(deriv, theta_init2, t, args=(g, ell))
theta3 = odeint(deriv, theta_init3, t, args=(g, ell))
theta4 = odeint(deriv, theta_init4, t, args=(g, ell))
```

(b) Make plots of θ vs. t, ω vs. t, and ω vs. θ (the so-called "phase plot") for $θ_0$ = 0.95π. Look carefully at their behavior. What is going on in these figures?

```python
plt.figure(figsize=(9, 5), dpi=180)
plt.subplot(1, 1, 1)

th0 = 0.95*np.pi

def deriv1(theta,t, g, ell): # return derivatives of the array theta
    return np.array([ theta[1], \
                     -1.0*(g/ell)*np.sin(theta[0]) ])

theta_init = np.array([th0, omega0])
theta = odeint(deriv1, theta_init, t, args=(g, ell))

plt.plot(t,theta[:,0],'g') #take only the theta values.
plt.xlabel(r'$t [seconds]$',fontsize = 'xx-large')
plt.ylabel(r'$\theta [rad]$',fontsize = 'xx-large')
plt.grid()
plt.title(r'$\theta $ as a function of time', fontsize = 'xx-large')

plt.show()
```
