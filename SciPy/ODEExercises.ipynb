```python
def deriv(theta,t, g, ell):
    return np.array([ theta[1], \
                     -1.0*(g/ell)*np.sin(theta[0]) ])
# (omega = 0 = dtheta, theta = domega) 
```

```python
g = 9.81
ell = 0.5
omega0 = 0.0
th0 = np.array([0.0, 45.0*np.pi/180., 90.0*np.pi/180., 135.0*np.pi/180., 180.0*np.pi/180.])
t = np.arange(0.,10.+0.01,0.01)

theta_approx0 = lambda t: th0[0]*np.cos(np.sqrt(g/ell)*t)
theta_approx1 = lambda t: th0[1]*np.cos(np.sqrt(g/ell)*t)
theta_approx2 = lambda t: th0[2]*np.cos(np.sqrt(g/ell)*t)
theta_approx3 = lambda t: th0[3]*np.cos(np.sqrt(g/ell)*t)
theta_approx4 = lambda t: th0[4]*np.cos(np.sqrt(g/ell)*t)

theta_init0 = np.array([th0[0],omega0])
theta_init1 = np.array([th0[1],omega0])
theta_init2 = np.array([th0[2],omega0])
theta_init3 = np.array([th0[3],omega0])
theta_init4 = np.array([th0[4],omega0])

theta0 = odeint(deriv, theta_init0, t, args=(g, ell))
theta1 = odeint(deriv, theta_init1, t, args=(g, ell))
theta2 = odeint(deriv, theta_init2, t, args=(g, ell))
theta3 = odeint(deriv, theta_init3, t, args=(g, ell))
theta4 = odeint(deriv, theta_init4, t, args=(g, ell))


#theta_init0 = np.zeros(2,1)
#for i in th0:
#    theta_init0[0,0] = i
#    print theta_init0
```

rtol, atol : float, optional

        The input parameters `rtol` and `atol` determine the error
        control performed by the solver.  The solver will control the
        vector, e, of estimated local errors in y, according to an
        inequality of the form ``max-norm of (e / ewt) <= 1``,
        where ewt is a vector of positive error weights computed as
        ``ewt = rtol * abs(y) + atol``.
        rtol and atol can be either vectors the same length as y or scalars.
        Defaults to 1.49012e-8.

$$\frac{e}{ewt} \leq 1 \tag{Max-norm}$$

$$ewt = rtol * |y| + atol$$

```python
m = 1.
g = 9.81
x0 = 0
y0 = 0
t = np.arange(0., 200.+0.1,0.1)
vter = 44

def derivative(y, t, m, g):
    vx = y[0]
    x1 = y[1]
    vy = y[2]
    y1 = y[3]
    dvxdt = -g/m * vx/vter**2 * np.sqrt(vx**2+vy**2)
    dvydt = -g/m * (1 + vy/vter**2 * np.sqrt(vx**2+vy**2))
    return np.array([dvxdt, vx, dvydt, vy])
```

```python
vx0 = 30*np.cos(50./180.*np.pi)
vy0 = 30*np.sin(50./180.*np.pi)

solve_by_odeint = odeint(derivative, np.array([vx0,x0,vy0, y0, vter]),t, args=(m,g))

vx = solve_by_odeint[:,0]
vy = solve_by_odeint[:,2]
x = solve_by_odeint[:,1]
y = solve_by_odeint[:,3]
```

```python
plt.figure(figsize=(9, 6), dpi=80)
plt.subplot(1, 1, 1)
plt.grid()
plt.plot(t, vx, 'r-', lw = 2, label = 'Velocity in X-D')
plt.plot(t, vy, 'g-', lw = 2, label = 'Velocity in Y-D')

plt.legend(loc = 'center right', fontsize = 'xx-large')
plt.xlabel("Velocity [meter/secs]",fontsize = 20)
plt.ylabel("time [secs]",fontsize = 20)
plt.title('Projectile', fontsize = 22)
plt.show()


plt.figure(figsize=(9, 6), dpi=80)
plt.subplot(1, 1, 1)
plt.grid()
plt.plot(x, y, 'r-', lw = 2, label = 'It is 2:30am. Wheee!')
plt.xlim(0,100)
plt.ylim(-10,40)
plt.legend(loc = 'upper right', fontsize = 'xx-large')
plt.xlabel("X-Displacement [meters]",fontsize = 20)
plt.ylabel("Y-Displacement [meters]",fontsize = 20)
plt.title('Projectile Path', fontsize = 22)
plt.show()
```

(b) Now create an interact object that uses sliders for the initial speed, terminal velocity, launch angle, and time. Have it plot two curves - one representing the solution with quadratic drag and the other in the absence of drag:

```python
from IPython.html.widgets import interact, interactive
```

```python
def proj(v_naught, v_terminal, theta, duration, y_naught):
    vx0 = v_naught * np.cos(theta*np.pi/180.)
    vy0 = v_naught * np.sin(theta*np.pi/180.)
    x_naught = 0  
    t = np.arange(0.,duration+0.1,.1)
    initialize = np.array([vx0,x_naught,vy0,y_naught,v_terminal])
    solve_by_odeint = odeint(derivative, initialize, t, args = (m, g))
    
    vx = solve_by_odeint[:,0]
    vy = solve_by_odeint[:,2]
    x = vx * t
    y = vy * t
    xt = x_naught + vx * t
    yt = y_naught + vy * t - .5 * g * t**2

    
    plt.figure(figsize=(14, 9), dpi=30)
    plt.subplot(1, 1, 1)
    plt.plot(x, y, label="This object is aerodynamically smooth.")
    plt.plot(xt, yt, label="This object was designed by an engineer. (with drag")
    plt.xlabel("X-Displacement")
    plt.ylabel("X-Displacement")

    plt.legend()
    plt.xlim(0,100)
    plt.ylim(0,80)
    plt.title('Projectile Path')
    plt.grid()
    plt.show()
    

v = interact(proj, theta=(0.,90.,1.), v_naught=(0.,100.,1.), v_terminal=(0.,110.,1.), duration=(0.,200.,1.), y_naught=(0,1.5,.1))
```
