### The Simple Pendulum as an example.

$$\sum \vec F = \vec T + \vec w = m \vec a $$

Newton's second law to sum the forces on the mass, $m$, of a pendulum. 

$$F_{radial} = T - mgcos(\theta) = ma_{radial} = 0 \tag{1}$$
$$F_{tangential} = -mgsin(\theta) = ma_{tangential} \tag{2}$$

$$s = l\theta \tag{3}$$
$$v_{tangential} = \frac{ds}{dt} = l \frac{d\theta}{dt}\tag{4}$$
$$a_{tangential} = \frac{d^2s}{dt^2} = l \frac{d^2 \theta}{dt^2}\tag{5}$$
$$\frac{d^2 \theta}{dt^2} = - \frac{g}{l} sin( \theta)\tag{6}$$

Since the length of the pendulum is fixed, there is no radial acceleration and so we only need to consider the accelerations tangential to the path of the mass. Using cylindrical coordinates, we find equations $(1)$ and $(2)$. The linear acceleration along the tangential direction is related to $\theta$ through the arc-length $s$, and so we see equation $(3)$ that is substituted into the tangential velocity and tangential acceleration as per equations $(4)$ and $(5)$. 

Since there is a $sin(\theta)$, we can see there is no analytic solution and so we must solve by approximation or numerically. 

### Approximate solution

Maclaurin series: 

$$sin(\theta) = \theta - \frac{\theta ^3}{3!} + \frac{\theta ^5}{5!} - \frac{\theta ^7}{7!} + ...\tag{7}$$

$$\longrightarrow \frac{d^2 \theta}{dt^2} = - \frac{g}{l} \theta \tag{8}$$

Which has solution: 

$$\theta (t) = \theta_0 cos(\sqrt{g l L }t)\tag{9}$$

```python
g = 9.8 #meters/sec^2
ell = 0.5 #meters (length)
th0 = 20.*np.pi/180 # 20 degrees, but must use radians. theta

t = np.arange(0.,10.+0.01,0.01) # 0 to 10.01 in steps of .01 (x-axis time)
theta_app = lambda t: th0*np.cos(np.sqrt(g/ell)*t) # equation 9.

plt.plot(t,theta_app(t),label='Approximate solution') # plot theta as a function of time. 
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
plt.ylim(-1,1.5)
plt.grid()
plt.legend()
plt.show()
```

This is only good for small angles < 50 degrees.

### Numerical Solution 

$\omega = \frac{d \theta}{ dt} \tag{angular velocity or angular frequency}$

$$\frac{d^2 \theta}{dt^2} = \frac{d \omega}{dt} = - \frac{g}{l} sin(\theta) \tag{10}$$

$$ \left[ \frac{d \theta}{dt} = \omega , \frac{d \omega}{dt} = - \frac{d}{l} sin(\theta) \right] \tag{11}$$

$$theta = \left[ \frac{d theta[0]}{dt} = theta[1], \frac{d theta[1]}{dt} = - \frac{g}{l}sin(theta[0]) \right] \tag{12}$$

```python
def deriv(theta_current,t):  # return derivatives of the array theta
    g = 9.8 #m/s^2  gravity constant.
    ell = 0.5 #m    length of pendulum.
    theta = theta_current[0] #
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([ dtheta, domega ])
```

```python
from scipy.integrate import odeint
```

```python
t = np.arange(0.,10.+0.01,0.01) # x-axis for plotting. 

omega0 = 0. #starting from rest, omega = 0 
theta_init = np.array([th0, omega0]) # initial values, th0 from before = 20 degrees, expressed in radians

#pass the function deriv to the solver with initial values and integration variable
theta = odeint(deriv,theta_init,t) 
```

```python
plt.plot(t,theta_app(t),'b-',label="Approximate solution")
plt.plot(t,theta[:,0],'r--',label="Numerical solution")
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
plt.ylim(-1,1.5)
plt.grid()
plt.legend()
plt.show()
```

```python
th0 = 50.*np.pi/180 # 50 degrees

theta_init = np.array([th0, omega0]) # initial values
theta = odeint(deriv,theta_init,t)

plt.plot(t,theta_app(t),'b-',label="Approximate solution")
plt.plot(t,theta[:,0],'r--',label="Numerical solution")
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
plt.ylim(-1.,1.5)
plt.grid()
plt.legend()
plt.show()
```

### Making physics come alive

```python
from IPython.html.widgets import interact, interactive
```

```python
def derivSP(thetaIn,t,g,ell):  # return derivatives of the array theta
    theta = thetaIn[0]
    omega = thetaIn[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return array([ dtheta, domega ])


def simplePendulum(th0=20.,w0=0.,ell=0.5,g=9.8):

    t = np.arange(0.,10.+0.01,0.01)

    #Approximate analytic solution for comparison
    approx = th0*np.pi/180*np.cos(np.sqrt(g/ell)*t)
  
    #convert degrees to radians
    initVals = np.array([th0*np.pi/180,w0])

    #pass the function deriv to the solver with initial values and integration variable
    theta = odeint(derivSP,initVals,t,args=(g,ell)) 
    #theta[:,0] are the positions
    #theta[:,1] are the angular velocities
    
    #Plot the results
    plt.plot(t,approx,label='Approximate solution')
    plt.plot(t,theta[:,0],label='Numerical solution')
    plt.xlabel(r'$t (seconds)$',fontsize=20)
    plt.ylabel(r'$\theta (rad)$',fontsize=20)
    #plt.ylim(-2.,2.)
    plt.grid()
    plt.legend()
    plt.show()

    
v = interact(simplePendulum,th0=(0.,90.),w0=(0.,10.),ell=(0.1,1.0),g=(1.0,20.0))
```

### Breaking into the black box

In order to solve for the motion of the simple pendulum using odeint, we convert the single second-order ODE into two first-order ODEs

#### Euler's Method

#### *Free-fall as an example

$$a(t) = \frac{d^2 y}{dt^2} \tag{1}$$ 

**Or** as two first-order DE:

$$a(t) = \frac{dv}{dt}, v(t) = \frac{dy}{dt} \tag{2}$$

$$v(t_1) = v(t_0) + a(t_0)\delta t \tag{3}$$

and for position:

$$y(t_1) = y(t_0) + v(t_0)\delta t \tag{4}$$

$$\longrightarrow \left[ v_{i+1} = v_i + a_i\Delta t, y_{i+1} = y_i + v_i\Delta t \right] \tag{5}$$

```python
def EulerFreefall():
    tmax = 10 # maximum time
    dt = 0.1 # time step
    a = -9.81 #g in m/s^2
    nsteps = int(tmax/dt)
    t = np.zeros(nsteps)
    v = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    # set initial conditions, redundant in this example
    t[0] = 0.
    v[0] = 0.

    for i in range(len(t)-1):
        t[i+1] = t[i] + dt
        v[i+1] = v[i] + dt * a
        y[i+1] = y[i] + dt * v[i]
        
    return t,v,y

te,ve,ye = EulerFreefall()
```

```python
plt.plot(te,ye,"r-",label="y(t)")
plt.plot(te,ve,"b-",label="v(t)")
plt.xlabel("time (s)",fontsize=20)
plt.ylabel("y(t),v(t)",fontsize=20)
plt.grid()
plt.legend(loc="best", fontsize=20)
plt.show()
```

### Exploring the parameter space

```python
def freefall(tmax=10.,dt=0.1,t0=0.,v0=46.,y0=0.):
    a = -9.81 #g in m/s^2

    #Set up the arrays based on the time parameters
    t = np.zeros(int(tmax/dt))
    v = np.zeros(int(tmax/dt))
    y = np.zeros(int(tmax/dt))

    #set up initial conditions from the parameters
    t[0] = t0
    v[0] = v0
    y[0] = y0

    #Apply Euler's method to solve
    for i in range(len(t)-1):
        t[i+1] = t[i] + dt
        v[i+1] = v[i] + dt * a
        y[i+1] = y[i] + dt * v[i]
    
    #Compare to the analytical result for ymax
    tmax = -v0/a
    ymax = y0 + v0*tmax + 0.5*a*tmax**2
    percentDiff = 100.*np.abs(y.max()-ymax)/ymax
    print "ymax percentDiff = %.2f"%percentDiff
    #plot it
    plt.plot(t,y,"r-",label="y(t)")
    plt.plot(t,v,"b-",label="v(t)")
    plt.xlabel("time (s)",fontsize=20)
    plt.ylabel("y(t),v(t)",fontsize=20)
    plt.xlim(0,20.)
    plt.ylim(-100.,600.)
    plt.grid()
    plt.legend(loc="upper left", bbox_to_anchor=(1,1), fontsize=20)
    plt.show()
```

```python
v4 = interact(freefall,tmax=(1.,20.),dt=(0.01,1.,0.01),t0=(0.,20.),v0=(10.,100.),y0=(0.,10.0))
```

### Euler-Cromer Method

Using equations 5:

```python
def EulerCromerFreefall():
    tmax = 10 # maximum time
    dt = 0.1 # time step
    a = -9.81 #g in m/s^2
    nsteps = int(tmax/dt)
    t = np.zeros(nsteps)
    v = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    # set initial conditions, redundant in this example
    t[0] = 0.
    v[0] = 0.

    for i in range(len(t)-1):
        t[i+1] = t[i] + dt
        v[i+1] = v[i] + dt * a
        y[i+1] = y[i] + dt * v[i+1] #use just calculated v value to compute y
        
    return t,v,y
```

```python
tec,vec,yec = EulerCromerFreefall()
```

```python
plt.plot(tec,yec,"r-",label="y(t)")
plt.plot(tec,vec,"b-",label="v(t)")
plt.xlabel("time (s)",fontsize=20)
plt.ylabel("y(t),v(t)",fontsize=20)
plt.grid()
plt.title("Euler-Cromer")
plt.legend(loc="best", fontsize=20)
plt.show()
```
