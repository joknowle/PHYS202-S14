```python
import numpy as np
def pointPotential(x,y,q,Xc,Yc):
    """
    return the electric potential for a point charge q at (Xc,Yc)
    Units are [Volts] if input is [meters][Coulombs]
    """
    k = 8.987551787997912e9 #(Nm^2/C^2)                                                                        
    Vxy = k*q/np.sqrt(((x-Xc)**2 + (y-Yc)**2))
    return Vxy
def dipolePotential(x,y,q,d):
    """
    Return electric potential for pair of point charges -/+ q
    separated by distance d along x axis with center (0,0)
    Units are [Volts] if inputs are [meters][Coulombs]
    """
    Vxy = pointPotential(x,y,q,-d/2.,0.) + pointPotential(x,y,-q, +d/2.,0.)
    return Vxy
```

### Question 2

Use the sample mean method to compute the volume of a sphere with radius 1 using spherical coordinates:

$$V = \int_0^{1}\int_0^{\pi}\int_0^{2\pi}r^2 sin(\theta)dr d\theta d\phi$$

for $N = (10^1,10^2,10^3,10^4,10^5,10^6)$. Report the value and statistical uncertainty for each sample size. How does the result depend on N?

```python
def sphere(N, r_max = 1., r_min = 0., theta_max = np.pi, theta_min = 0., phi_max = 2. * np.pi, phi_min = 0.):
    
    #set up the function
    func = lambda r, theta, phi: r**2*np.sin(theta) #function for integration.
    
    r_s = (r_max - r_min) * np.random.random_sample(N)
    theta_s = (theta_max - theta_min) * np.random.random_sample(N)
    phi_s = (phi_max-phi_min)*np.random.random_sample(N)
    error = (r_max) * (theta_max) * (phi_max) * np.sqrt((( \
                    func(r_s, theta_s, phi_s)**2).mean() - (func(r_s, theta_s, phi_s).mean())**2)/ N)
    approx = theta_max * phi_max * func(r_s, theta_s, phi_s)
    I = approx.mean()
    print "N=", N, "I =", I, "+/-", error
    
sphere(N = 10) 
sphere(N = 100)
sphere(N = 1000) 
sphere(N = 10000)
sphere(N = 100000) 
sphere(N = 1000000) 
```

As $N$ increases, the error decreases, so there is a proportional relationship.

### Question 3

(a) Use the hit or miss Monte Carlo method to estimate the integral of $f(x) = 4 \sqrt{1-x^2}$ in the interval $0 \leq x \leq 1$ as a function of $N$, the number of samples. Let $N$ range from $10^1$ to $10^6$ and use at least $100$ different values for $N$.
(np.logspace might be useful here to get evenly spaced values for N.)

```python
def hitOrMiss(N = 10., x_min = 0., x_max = 1.): # Monty Carlo hit or miss method
    func = lambda x: 4. * sqrt(1. - x**2.)
    x = np.arange(-.5,1.5,0.001)
    subx = x[logical_and((x > x_min),(x < x_max))]
    y_max = func(subx).max() * 1.05
    y_min= 0.
    for i in np.logspace(1.,6.,6.):
        #integration
        x = (x_max - x_min) * np.random.random_sample(i)
        y = (y_max - y_min) * np.random.random_sample(i)
        mask = (y < func(x))
        
        I = x_max * y_max * mask.sum()/mask.size
        print "I =", I, "when N = ", i
hitOrMiss()
```

(b) Calculate the percent error for the $N$ values from part (a) using the exact result $\pi$. Make a log-log plot of the error as a function of $N$ .
What is the approximate functional dependence of the error on $N$  for large $N$ , for example $N \geq 10^4$?  

```python
def hitOrMiss2(N = 10., x_min = 0., x_max = 1.):
    func = lambda x: 4. * np.sqrt(1. - x**2.)
    x = np.arange(-.5,1.5,0.001)
    exact_I = np.pi
    N_array = []
    err_I = []
    subx = x[logical_and((x > x_min),(x < x_max))]
    y_min= 0.
    y_max = func(subx).max() * 1.05
    for i in np.logspace(1., 6., 150.):
        x = (x_max - x_min) * np.random.random_sample(i)
        y = (y_max - y_min) * np.random.random_sample(i)
        mask = (y < func(x))
        I = x_max * y_max * mask.sum() / mask.size
        
        error = 100. * np.abs(exact_I - I) / exact_I
        err_I.append(error)
        N_array.append(i)
        
    plt.grid()
    plt.yscale('log')
    plt.xscale('log')
    plt.title("Error as a function of Cells")
    plt.ylabel("Error", fontsize = 16)
    plt.xlabel("Cells", fontsize = 16)
    plt.plot(N_array, err_I, 'o', color = 'red')
    plt.show()
hitOrMiss2()
```

There exists a somewhat linear relationship between the error of $N$ for large $N$.

(c) Estimate the same integral using the sample mean method and compute the statistical uncertainty and error as a function of the number of samples $N$ for $N \geq 10^4$. Use at least $100$ different values for $N$.
Make a log-log plot of both the statistical uncertainty and the error as a function of $N$.
How many samples are needed to determine the result to two decimal places? What is the approximate functional dependence of the error on $N$ for large $N$?

```python
def sampleMeanAnalysis(N, x_min = 0., x_max = 1.):

    func = lambda x: 4. * np.sqrt(1. - x**2.)
    uncertainty = []
    exact_I = np.pi
    N_array = []
    err_I = []
    
    for i in N:
        x = (x_max - x_min) * np.random.random_sample(i)
        appx = x_max * func(x)
        I = appx.mean()
        error = 100. * np.abs(I - exact_I) / exact_I
        newUncertainty = x_max * np.sqrt(((func(x)**2.).mean() - (func(x).mean())**2.)/i)
        err_I.append(error)
        uncertainty.append(newUncertainty)   
        N_array.append(i)
        
    plt.title("Error as a function of Cells")
    plt.yscale('log')
    plt.xscale('log')
    plt.grid()
    plt.ylabel("Error")
    plt.xlabel("Cells")
    plt.plot(N_array, uncertainty, 'ro', label='Uncertainty')
    plt.plot(N_array, err_I, 'go', label="Error")
    plt.legend(loc='upper right')
    plt.show()
    
x = np.logspace(4., 7., 100.)
sampleMeanAnalysis(x)
```

There must be over $10^6$ cells to produce an accuracy of two sig figs with $\approx 50$%. 
There is a linear relationship between the error of $N$ for large $N$.
